```python
mhmm = numpy.array([])
for row in OPzone_data.values:
    if row[-1]==1 and row[-2]==1:
        mhmm=numpy.append(mhmm,3)
        continue
    if row[-1]==1:
        mhmm=numpy.append(mhmm,1)
        continue
    if row[-2]==1:
        mhmm=numpy.append(mhmm,2)
        continue
    else:
        mhmm=numpy.append(mhmm,0)
    
        
OPzone_data['OP_QCT']=mhmm
```

```python
import json
from bokeh.io import show
from bokeh.models import (CDSView, ColorBar, ColumnDataSource,
                          CustomJS, CustomJSFilter, 
                          GeoJSONDataSource, HoverTool,
                          LinearColorMapper, Slider,Legend,LegendItem)
from bokeh.layouts import column, row, widgetbox
from bokeh.palettes import brewer
from bokeh.plotting import figure
```

```python
# Input GeoJSON source that contains features for plotting
geosource = GeoJSONDataSource(geojson = gdf.to_json())
# Define color palettes
palette = brewer['Set2'][8]
palette = palette[::-1] # reverse order of colors so higher values have darker colors

# Instantiate LinearColorMapper that linearly maps numbers in a range, into a sequence of colors.
color_mapper = LinearColorMapper(palette = palette, low = 0, high = 3)

# Create color bar.
color_bar = ColorBar(color_mapper = color_mapper, 
                     label_standoff = 8,
                     width = 500, height = 20,
                     border_line_color = None,
                     location = (0,0), 
                     orientation = 'horizontal')

p = figure(title = 'QCT and Opportunity Zones 2020', 
           plot_height = 600, plot_width = 950, 
           toolbar_location = 'below',
           tools = 'pan, wheel_zoom, box_zoom, reset')
p.xgrid.grid_line_color = None
p.ygrid.grid_line_color = None

# Add patch renderer to figure.
states = p.patches('xs','ys', source = geosource,
                   fill_color = {'field' :'OP_QCT',
                                 'transform' : color_mapper},
                   line_color = 'gray', 
                   line_width = 0.25, 
                   fill_alpha = 1)
# Create hover tool
p.add_tools(HoverTool(renderers = [states],
                      tooltips = [('State ','@State_name'),
                                  ('County ','@cntyname'),
                                  ('Tract_id ','@tract_id'),
                                  ('Average TAHSAIL Ratio','@Average_TAHSAIL_Tract_Ratio'),
                                  ('Average Poverty Rate','@Average_Poverty_Rate'),
                                ('QCT','@QCT_2021'),('Opportunity Zone','@Opportunity_Zone')]))


legend = Legend(items=[
    LegendItem(label="Opportunity Zone", renderers=[states], index=5),
    LegendItem(label="QCT", renderers=[states], index=1),
    LegendItem(label="Both", renderers=[states], index=4)
])
p.add_layout(legend)

show(p)

```
